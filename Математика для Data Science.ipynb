```python
def f(x, a, b):
    return np.e**a * np.sin(b*x) + x
```

```python
def error(params):
    return np.sum(np.abs(fx - f(x, params[0], params[1])))
```

```python
a_range = np.arange(0, 2, 0.1)
b_range = np.arange(0, 2, 0.1)

# Инициализируем переменные для хранения лучших параметров и RES
best_params = None
best_res = float('inf')

# Перебираем параметры и находим модель с наименьшим RES
for a in a_range:
    for b in b_range:
        # Вызываем curve_fit для текущих параметров
        params, _ = curve_fit(f, x, fx, [a, b])

        # Вычисляем прогнозные значения и RES
        fx_pred = f(x, *params)
        res = error(params)

        # Сохраняем параметры, если RES меньше текущего лучшего
        if res < best_res:
            best_params = params
            best_res = res

# Выводим лучшие параметры и RES
print('Best parameters:', best_params)
print('Best RES:', best_res)

plt.figure(figsize = (12, 8))
plt.plot(x, fx, color='blue')
plt.plot(x, f(x, best_params[0], best_params[1]), color = 'red', linestyle = '--', linewidth = 5)
plt.show()
```

При фиксированном `seed=42` поэкспериментируйте с параметрами алгоритма [differential_evolution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.differential_evolution.html): strategy, popsize, mutation.

Постройте графики количества итераций (`nit`) оптимизации [функции ackley](https://en.wikipedia.org/wiki/Ackley_function) от значения параметра.

Подробнее о результате выдачи [тут](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.OptimizeResult.html).

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]

result = differential_evolution(ackley, bounds, seed=42)
result
```

```python
stg =['best1bin', 'best2exp', 'best1exp', 'randtobest1exp', 'rand1bin']
nit = []
for i in stg:
  res = differential_evolution(ackley,  bounds,  strategy= i,  seed=42 )['nit']
  nit.append(res)
nit
```

```python
plt.figure(figsize = (12, 8))
plt.plot(stg, nit, color='blue')
plt.show()
```

```python
psz =[1, 2, 3, 5, 8]
nit = []
for i in psz:
  res = differential_evolution(ackley,  bounds,  popsize= i,  seed=42 )['nit']
  nit.append(res)
nit
```

```python
plt.figure(figsize = (12, 8))
plt.plot(psz, nit, color='green')
plt.show()
```

```python
mut =[0.1, 0.9, 1.0, 1.2, 1.9]
nit = []
for i in mut:
  res = differential_evolution(ackley,  bounds,  mutation= i,  seed=42 )['nit']
  nit.append(res)
nit
```

```python
plt.figure(figsize = (12, 8))
plt.plot(mut, nit, color='red')
plt.show()
```

## Косинусное расстояние векторов. Поиск наименьшего значения



Дана матрица рейтингов фильмов (строки - пользователи (10 человек), столбцы - фильмы (15 фильмов))

```python
raitings = np.array([[ 4,  4,  9,  4,  1,  6, 10,  7,  9,  6,  9,  2,  8,  6,  6],
                     [ 9,  2,  5, 10,  7,  8, 10,  5,  6,  2,  1,  6,  8,  9,  7],
                     [ 1,  6,  8,  8,  4,  9,  3,  8, 10,  5,  2,  6,  8,  1,  6],
                     [ 6,  1,  9,  7,  7,  9,  2,  3,  5,  1,  6,  6,  3,  2,  7],
                     [ 3,  7,  3,  5,  7,  9,  9,  6,  2,  9,  1,  2,  8, 10,  6],
                     [ 8,  3,  7,  3,  8,  6,  1,  8,  8,  6,  1,  9,  4, 10,  1],
                     [ 9,  8,  4,  8,  8, 10,  6,  1,  1,  2,  9,  5,  2,  7,  2],
                     [ 4,  1,  6,  4,  3, 10,  4,  4,  2,  8,  7,  9,  3,  8,  3],
                     [ 2,  7,  7,  6, 10,  6,  8,  9,  8,  6, 10,  1,  7, 10,  4],
                     [ 5, 10,  8,  8,  9,  7,  2,  9,  9, 10,  8,  8,  8,  6, 10]])
```

Необходимо найти наиболее похожего пользователя по косинусной метрике для каждого из пользователей 

```python
# определим функцию для расчета косинусного сходства между двумя векторами
def cosine(a, b):
    # определим длины векторов
    a_len = np.linalg.norm(a)
    b_len = np.linalg.norm(b)
    # косинусным сходством будет являться частное от деления скалярного произведения векторов на произведение длин векторов
    return np.dot(a,b) / (a_len*b_len)
```

```python
# Матрица результатов косинусных расстояний
cosine_dist = np.zeros((10,10))
```

```python
# Переберем строки массива и посчитаем косинусные расстояния между всеми векторами
for i in range(len(raitings)):
    a = raitings[i]
    for j in range(len(raitings)):
        b = raitings[j]
        cosine_ = cosine(a, b)
        if i==j: #обнулим сходство элементов на главной диагонали результирующей матрицы, 
            # чтобы они не участвовали в сравнении сами с собой (иначе их сходство будет равно максимальному, т.е 1.0)
            cosine_ = 0 
        cosine_dist[i][j] = cosine_
        
cosine_dist
```

```python
for i in range(len(cosine_dist)):
    print("Для пользователя", i+1, "наиболее похожий пользователь:", np.argmax(cosine_dist[i])+1)
```
